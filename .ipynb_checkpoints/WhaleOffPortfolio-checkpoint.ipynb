```python
# Prep all the pathways
import pandas as pd
import numpy as np
import datetime as dt
from pathlib import Path
%matplotlib inline
```

```python
# Use Pandas to read csv as a Dataframe for analysis of Whale returns.
whale_returns_csv = Path("Starter_code/Resources/whale_returns.csv")
whale_returns = pd.read_csv(whale_returns_csv, index_col="Date", infer_datetime_format=True, parse_dates=True)
# whale_returns.describe()
whale_returns.sort_index(ascending=True, inplace=True)
whale_returns.head()
```

```python
# Count, detect and remove the null
whale_returns.isnull().sum()
```

```python
# Drop nulls
whale_returns.dropna(inplace=True)
whale_returns.isnull().sum()
```

```python
# Reading algorithmic returns and clean the data
algo_returns_csv = Path("Starter_Code/Resources/algo_returns.csv")
algo_returns = pd.read_csv(algo_returns_csv, index_col="Date", infer_datetime_format=True, parse_dates=True)
algo_returns.sort_index(ascending=True, inplace=True)
algo_returns.head()
```

```python
# Count nulls
algo_returns.isnull().sum()
```

```python
# Drop nulls
algo_returns.dropna(inplace=True)
algo_returns.head()
```

```python
# Reading algorithmic returns
sp_history_csv = Path("Starter_Code/Resources/sp_history.csv")
sp_history = pd.read_csv(sp_history_csv, index_col="Date", infer_datetime_format=True, parse_dates=True)
sp_history.sort_index(ascending=True, inplace=True)
sp_history.head()
```

```python
# Check data types
sp_history.dtypes
```

```python
# Fix data types
sp_history["Close"]= sp_history["Close"].str.replace('$', '').astype("float")
sp_history.sort_index(inplace=True)
sp_history
```

```python
# Calculate daily returns
sp_daily_returns = sp_history.pct_change()
```

```python
# Drop nulls
sp_daily_returns.dropna(inplace=True)
sp_daily_returns.head()
```

```python
# Rename column
sp_daily_returns.rename(columns = {"Close": "S&P 500"}, inplace=True)
```

```python
# Combine Whale, Algorithmic, and S&P 500 returns by concatenating all Dataframes into a single Dataframe
df_daily = pd.concat([whale_returns, algo_returns, sp_daily_returns], axis = "columns", join = "inner")
df_daily
```

```python
# Calculate adn plot the daily returns
df_daily.plot(figsize = (20,10))
```

```python
# Plot cumulative returns
df_cum = (1+ df_daily).cumprod()
df_cum.plot(figsize = (20,10))
```

```python
# Box plot to visually show risk
df_daily.plot(kind = "box", figsize = (20,10))
```

```python
# Calculate the standard deviation for each portfolio. Which portfolios are riskier than the S&P500?
portfolio_risk = df_daily.std()
portfolio_risk
```

```python
# Determine which portfolios are riskier than the S&P 500 by using the Python apply.lambda function. 
SP_500_Risk = portfolio_risk[-1]
portfolio_risk.to_frame()
portfolio_risk.apply(lambda x : True if x> SP_500_Risk else False)
```

```python
# Calculate the annualized standard deviation (252 trading days)
annual_portfolio_risk = df_daily.std() * np.sqrt (252)
annual_portfolio_risk
```

```python
# ROLLING STATISTICS
# Calculate and plot the rolling standard deviation for the S&P 500 using a 21 day window
df_daily.rolling(window=21).std().plot(figsize=(20,10))
```

```python
# Correlation
price_correlation = df_daily.corr()
price_correlation
```

```python
# Calculate Beta for a single portfolio compared to the total market (S&P 500)
covariance = df_daily["BERKSHIRE HATHAWAY INC"].cov(df_daily["sp500"])
covariance
```

```python
variance = df_daily['BERKSHIRE HATHAWAY INC'].var()
variance
```

```python
berkshire_beta = covariance / variance
berkshire_beta
```

```python
rolling_covariance = df_daily['BERKSHIRE HATHAWAY INC'].rolling(window=60).cov(df_daily['sp500'])
rolling_variance = df_daily['sp500'].rolling(window=60).var()
```

```python
rolling_beta = rolling_covariance / rolling_variance
rolling_beta.plot(figsize=(20,10), title='Berkshire Hathaway Inc. Beta')
```

```python
sp500_rolling_21 = df_daily[["sp500"]].rolling(window = 21).std()
sp500_rolling_21.plot(figsize=(20,10))
```

```python
# Exponentially weighted average
df_daily.ewm(halflife=21).std().plot(figsize=(20,10))
```

```python
#Annualized Sharpe Ratios
sharpe_ratios = (df_daily.mean()* 252) / (df_daily.std() * np.sqrt(252))
sharpe_ratios
```

```python
# Visualize the Sharpe Ratios as a bar plot
sharpe_ratios.plot(kind="bar")
```

```python
# Choose 3-5 custom stocks with at last 1 year's worth of historic prices and create a DataFrame of the closing prices and dates for each stock.
# Read the first Stock
aapl_historical_csv = Path("Starter_Code/Resources/aapl_historical.csv")
aapl_historical = pd.read_csv(aapl_historical_csv, index_col="Trade DATE", infer_datetime_format=True, parse_dates=True)
aapl_historical.head()
```

```python
# Read the second stock
cost_historical_csv = Path("Starter_Code/Resources/cost_historical.csv")
cost_historical = pd.read_csv(cost_historical_csv, index_col="Trade DATE", infer_datetime_format=True, parse_dates=True)
cost_historical.head()
```

```python
# Read the third stock
goog_historical_csv = Path("Starter_Code/Resources/goog_historical.csv")
goog_historical = pd.read_csv(goog_historical_csv, index_col="Trade DATE", infer_datetime_format=True, parse_dates=True)
goog_historical.head()
```

```python
# Concatenate all stocks into a single Dataframe
custom_df = pd.concat([aapl_historical, cost_historical, goog_historical], axis="rows", join="outer")
custom_df.sort_index(ascending=True, inplace=True)
custom_df.head()
```

```python
#sort index
#custom_df.sort_index(inplace=True)
#custom_df.head()
```

```python
# Reset the index
custom_df.reset_index()
```

```python
# Pivot the data so that the stock tickers are the columns, the dates are the index, and the values are the closing price
custom_df = custom_df.pivot(columns="Symbol", values="NOCP")
custom_df.head()
```

```python
# Drop Nulls
custom_daily_returns = custom_df.pct_change()
custom_daily_returns.dropna(inplace=True)
custom_daily_returns.head()
```

```python
# Calculate weighted portfolio returns
weights = [1/3, 1/3, 1/3]
custom_portfolio_returns = custom_daily_returns.dot(weights)
custom_portfolio_returns.head()
```

```python
all_portfolio_returns = pd.concat([df_daily, custom_portfolio_returns], axis="columns", join="inner")
x = all_portfolio_returns.columns[-1]
all_portfolio_returns = all_portfolio_returns.rename(columns={x:"Custom"})
all_portfolio_returns.columns
```

```python
all_portfolio_returns.isnull()
all_portfolio_returns.dropna(axis=0, how="any", inplace=True)
all_portfolio_returns
```

```python
# Risk
all_portfolio_risk = all_portfolio_returns.std() * np.sqrt (252)
all_portfolio_risk
```

```python
all_portfolio_returns.rolling(window=21).std().plot(figsize=(20,10), title="21 Day Rolling Standard Deviation")
```

```python
# Beta
covariance_all = all_portfolio_returns['Custom'].cov(all_portfolio_returns['sp500'])
covariance_all
variance_all = all_portfolio_returns['sp500'].var()
berkshire_beta = covariance / variance
rolling_covariance_all = all_portfolio_returns['Custom'].rolling(window=60).cov(all_portfolio_returns['sp500'])
rolling_variance_all = all_portfolio_returns['sp500'].rolling(window=60).var()
rolling_beta_all = rolling_covariance_all / rolling_variance_all
rolling_beta_all.plot(figsize=(20, 10), title='Custom Portfolio Beta')
```

```python
# Annualzied Sharpe Ratios
sharpe_ratios_all = (all_portfolio_returns.mean() * 252) / (all_portfolio_returns.std() * np.sqrt(252))
sharpe_ratios_all
```

```python
# Visualize the sharpe ratios as a bar plot
sharpe_ratios_all.plot(kind="bar")
```
